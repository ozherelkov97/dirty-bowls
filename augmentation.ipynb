```python
import numpy as np
import os
from PIL import Image, ImageEnhance  
path = os.getcwd()
import shutil

def resize(im, n, m):
    resized = im.resize((n, m), Image.ANTIALIAS)
    return resized

def resize_folder(direct_in, direct_out, n, m):
    if os.path.exists(direct_out):
        shutil.rmtree(direct_out)
    if not os.path.exists(direct_out):
        os.makedirs(direct_out)
    files=os.listdir(direct_in)
    for im in files:
        image=Image.open(direct_in+'/'+im)
        image = resize(image, n, m)
        image.save(direct_out+'/'+im)
        
def augmentation(direct_in, direct_out):
    
    if os.path.exists(direct_out):
        shutil.rmtree(direct_out)
    if not os.path.exists(direct_out):
        os.makedirs(direct_out)
        
    files=os.listdir(direct_in)
    for im in files:
        image=Image.open(direct_in+'/'+im)
        #image = resize(image, 256, 256)
        image.save(direct_out+'/'+im+'1.jpg')
        img_dark = ImageEnhance.Brightness(image).enhance(0.3)
        img_dark = ImageEnhance.Contrast(img_dark).enhance(3.0)
        img_dark.save(direct_out+'/'+im+'2.jpg')
        img_light = ImageEnhance.Brightness(image).enhance(1.3)
        img_light = ImageEnhance.Contrast(img_light).enhance(2.0)
        img_light.save(direct_out+'/'+im+'3.jpg')
        image_trans=image.transpose(Image.FLIP_LEFT_RIGHT)
        image_trans.save(direct_out+'/'+im+'4.jpg')
        img_dark = ImageEnhance.Brightness(image_trans).enhance(0.3)
        img_dark = ImageEnhance.Contrast(img_dark).enhance(3.0)
        img_dark.save(direct_out+'/'+im+'5.jpg')
        img_light = ImageEnhance.Brightness(image_trans).enhance(1.3)
        img_light = ImageEnhance.Contrast(img_light).enhance(2.0)
        img_light.save(direct_out+'/'+im+'6.jpg')

        image=image.rotate(90)

        image.save(direct_out+'/'+im+'7.jpg')
        img_dark = ImageEnhance.Brightness(image).enhance(0.3)
        img_dark = ImageEnhance.Contrast(img_dark).enhance(3.0)
        img_dark.save(direct_out+'/'+im+'8.jpg')
        img_light = ImageEnhance.Brightness(image).enhance(1.3)
        img_light = ImageEnhance.Contrast(img_light).enhance(2.0)
        img_light.save(direct_out+'/'+im+'9.jpg')
        image_trans=image.transpose(Image.FLIP_LEFT_RIGHT)
        image_trans.save(direct_out+'/'+im+'10.jpg')
        img_dark = ImageEnhance.Brightness(image_trans).enhance(0.3)
        img_dark = ImageEnhance.Contrast(img_dark).enhance(3.0)
        img_dark.save(direct_out+'/'+im+'11.jpg')
        img_light = ImageEnhance.Brightness(image_trans).enhance(1.3)
        img_light = ImageEnhance.Contrast(img_light).enhance(2.0)
        img_light.save(direct_out+'/'+im+'12.jpg')


        image=image.rotate(90)

        image.save(direct_out+'/'+im+'13.jpg')
        img_dark = ImageEnhance.Brightness(image).enhance(0.3)
        img_dark = ImageEnhance.Contrast(img_dark).enhance(3.0)
        img_dark.save(direct_out+'/'+im+'14.jpg')
        img_light = ImageEnhance.Brightness(image).enhance(1.3)
        img_light = ImageEnhance.Contrast(img_light).enhance(2.0)
        img_light.save(direct_out+'/'+im+'15.jpg')
        image_trans=image.transpose(Image.FLIP_LEFT_RIGHT)
        image_trans.save(direct_out+'/'+im+'16.jpg')
        img_dark = ImageEnhance.Brightness(image_trans).enhance(0.3)
        img_dark = ImageEnhance.Contrast(img_dark).enhance(3.0)
        img_dark.save(direct_out+'/'+im+'17.jpg')
        img_light = ImageEnhance.Brightness(image_trans).enhance(1.3)
        img_light = ImageEnhance.Contrast(img_light).enhance(2.0)
        img_light.save(direct_out+'/'+im+'18.jpg')

        image=image.rotate(90)

        image.save(direct_out+'/'+im+'19.jpg')
        img_dark = ImageEnhance.Brightness(image).enhance(0.3)
        img_dark = ImageEnhance.Contrast(img_dark).enhance(3.0)
        img_dark.save(direct_out+'/'+im+'20.jpg')
        img_light = ImageEnhance.Brightness(image).enhance(1.3)
        img_light = ImageEnhance.Contrast(img_light).enhance(2.0)
        img_light.save(direct_out+'/'+im+'21.jpg')
        image_trans=image.transpose(Image.FLIP_LEFT_RIGHT)
        image_trans.save(direct_out+'/'+im+'22.jpg')
        img_dark = ImageEnhance.Brightness(image_trans).enhance(0.3)
        img_dark = ImageEnhance.Contrast(img_dark).enhance(3.0)
        img_dark.save(direct_out+'/'+im+'23.jpg')
        img_light = ImageEnhance.Brightness(image_trans).enhance(1.3)
        img_light = ImageEnhance.Contrast(img_light).enhance(2.0)
        img_light.save(direct_out+'/'+im+'24.jpg')
```

```python
input_dir = os.path.join(path, 'original', 'cleaned')
output_dir = os.path.join(path, 'augmentation', 'cleaned')
augmentation(input_dir, output_dir)

input_dir = os.path.join(path, 'original', 'dirty')
output_dir = os.path.join(path, 'augmentation', 'dirty')
augmentation(input_dir, output_dir)
```
